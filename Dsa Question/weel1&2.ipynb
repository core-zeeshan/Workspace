Description
Write a function two_sum(nums, target) that takes a list of integers nums and an integer target. Return the indices of
the two numbers that add up to target. You may assume exactly one solution exists, and you cannot use the same
element twice. Use a dictionary for O(n) time.
Example Input/Output

two_sum([2, 7, 11, 15],9)
[0, 1]
two_sum([3, 2, 4], 6)
[1, 2]


```python
def two_sum(nums,target):
    for i in range(len(nums)):
        for j in range(1,len(nums)):
            if nums[i]+nums[j] == target:
                return i,j
nums=[3,2,4]
target=6
two_sum(nums,target)
```

Problem 2 - Sort By Parity
Description
Given a list of integers nums, write a function sort_by_parity(nums) that returns a new list where all even numbers
appear before all odd numbers. The relative order of evens/odds among themselves can be arbitrary. Solve it using list
comprehension (two passes) or an in-place two-pointer approach.
Example Input/Output
 sort_by_parity([3, 1, 2, 4])
[2, 4, 3, 1] # or [4, 2, 1, 3] etc. - any valid parity order works.
 sort_by_parity([0, 1, 2, 3])
[0, 2, 1, 3]


```python
def sort_by_parity(nums):
    evens = [x for x in nums if x % 2 == 0]
    odds = [x for x in nums if x % 2 != 0]
    return evens + odds

print(sort_by_parity([3, 1, 2, 4]))
print(sort_by_parity([0, 5, 7, 8, 2]))
```

Problem 4 - Contains Duplicate II
Description
Given a list of integers nums and an integer k, write a function contains_nearby_duplicate(nums, k) that returns True
if there exist two distinct indices i and j such that nums[i] == nums[j] and the absolute difference between i and
j is at most k. Otherwise return False. Use a dictionary/hash map to achieve O(n) time.
Example Input/Output

contains_nearby_duplicate([1, 2, 3, 1], 3)

True 
indices 0 and 3 → difference 3 ≤3

contains_nearby_duplicate([1, 0, 1, 1], 1)

True 

indices 2 and 3 → difference 1 ≤1

contains_nearby_duplicate([1, 2, 3, 1, 2, 3], 2)

False

```python
def contain_nearby_duplicate(nums,k):
    for i in range(len(nums)):
        for j in range(1,len(nums)):
            if nums[i] == nums[j] and abs(i - j) <= k:
                return True
    return False
nums=[1, 2, 3, 1]
k=3
contain_nearby_duplicate(nums,k)
```
