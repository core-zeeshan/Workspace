Q: WAP tha read a comma separated values list of number from the users and store them in a listthen calculate and print the sum and average


Using  split() method.

```python

a = input("Give a comma-Separated Value like 1,2,3,4,5")

# print(type(a))
b =a.split(",")
# print(b)
x=[]
for i in b:
    x.append(int(i))
print(x)
sum =0
avg =0
for i in x:
    sum+=i
print(f"Sum:{sum}")
print(f"Average:{sum/len(x)}")
    
```

```python
a = input("Give a comma-Separated Value like 1,2,3,4,5")
print(f"User Given Data : {a}")
print(f"Data type of  user given data is {type(a)}")
x=[]
temp = ""
for i in a:
    if i != ",":
        temp += i
    else:
        x.append(int(temp))
        temp = ""

x.append(int(temp))

print(f"Changed user data into list:{x}")
print(f"Changed Data Type is {type(x)}")
sum=0
for i in x:
    sum+=i
print(f"Sum:{sum}")
print(f"Average:{round(sum/len(x),2)}")
```

Write a function of finding largest element without using max() function 


```python
a = [2,43,54,6,87,98,234,34,546]
def find_max(a):
    temp = 0
    for i in a:
        if i>= temp:
            temp=i
    return temp
print(find_max(a))
        
```

Q: vowel count lower case and upper case

```python
a= input("Enter your string to check vowel:")
count = 0
for i in a:
    if i == "a" or i == "A":
        count+=1
    elif i == "e" or i == "E":
        count+=1
    elif i == "i" or i == "I":
        count+=1
    elif i == "o" or i == "O":
        count+=1
    elif i == "u" or i == "U":
        count+=1    
print(f"Input:{a}")
print(f"No of vowel :{count}")
```

```python
a= input("Enter your string to check vowel:")
count = 0
lst = "aeiouAEIOU"
for ch in a:
    if ch in lst:
        count+=1

print(f"Input:{a}")
print(f"No of vowel :{count}")
```

WAP to make calculator make function of operation

```python
# opr = input("Enter Operator (+,-,*,/)")
# opd1 = input("First number:")
# opd2 = input("Second number:")

def add(opd1,opd2):
    return opd1+opd2
def subtract(opd1,opd2):
    return opd1-opd2
def multiply(opd1,opd2):
    return opd1*opd2
def divide(opd1,opd2):
        return opd1/opd2


try:
    opr = input("Enter operator (+, -, *, /): ")
    opd1 = float(input("Enter first number: "))
    opd2 = float(input("Enter second number: "))
    

    if opr == "+":
        result = add(opd1, opd2)
    elif opr == "-":
        result = subtract(opd1, opd2)
    elif opr == "*":
        result = multiply(opd1, opd2)
    elif opr == "/":
        result = divide(opd1, opd2)
    else:
        result = "Invalid operator"
    print("Result:", result)

except ZeroDivisionError:
    print("Error: Can't divide by 0")

except ValueError:
    print("Error: Please enter valid numbers")
```
