# Bubble Sort

```python
a = [8, 3, 15, 1, 9, 6, 12, 4, 7, 2]
length = len(a)
print(a)
for i in range(length-1):
    swap = False
    for j in range(length - 1-i):
        if a[j]>a[j+1]:
            print(f"{a[j]} is swapped with {a[j+1]}")
            a[j],a[j+1]=a[j+1],a[j]
            print(a)
            swap=True
        else:
            print(f"{a[j]} is not swapped with {a[j+1]} because {a[j]} is smaller than {a[j+1]} ")
            
    print("")
    print(f"Pass {i} : {a}")
    print("")
    if swap==False:
        print("Completed")
        break
    print(f"Now Pass {i+1}")
    
        
```

```python
a=[3,5,7,8,9,78]
length = len(a)
for i in range(length-1):
    swap=False
    print(f"Pass:{i}")
    for j in range(length-1-i):
        print(f"Pass{j}")
        
        if a[j]>a[j+1]:
            a[j],a[j+1]=a[j+1],a[j]
            print(f"Swapped Value this:{a[j+1]} to this {a[j]}")
            swap = True
        
    if swap == False:
        break
print(a)
```
