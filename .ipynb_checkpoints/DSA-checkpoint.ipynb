Coding problem 1: Check if an array contains any duplicate elements.
input: [1,2,3,1]
Output: True






```python
arr = [1,2,3,1]
duplicate = False
for i in range(len(arr)):
    for j in range(i+1,len(arr)):
        if arr[i] == arr[j]:
            duplicate = True

print(duplicate)
```

coding problem 2: Merge two sorted arrays into a single sorted array.
Input: [1,3,5], [2,4,6]
Output: [1,2,3,4,5,6]

```python
arr1 = [1,4,3,5]
arr2=[2,4,6]
merged_arr = []
for i in arr1:
    merged_arr.append(i)
for j in arr2:
    merged_arr.append(j)

print(merged_arr)


for i in range(len(merged_arr)):
    for j in range(i+1,len(merged_arr)):
        if merged_arr[i]>merged_arr[j]:
            temp = merged_arr[i]
            merged_arr[i] = merged_arr[j]
            merged_arr[j] = temp
       
print(f"Sorted Array:{merged_arr}")
```

 coding problem2: Problem: Return the count of even integers in a given list.






```python
a = [5,67,7,25,6,5,23,5,7,8,9,4,12,78]
def even_int(a):

    count = 0
    for i in a:
        if i % 2==0:
            count=count+1
    return count

print(even_int(a))
```

comding problem1: Problem: Given a list of integers (can have duplicates), find the second largest distinct element. 
If it doesn't exist (e.g., list has less than 2 unique values), return None.




```python
lst = [2,3,5,2,3,6,2,8,75,34]
def second_largest(lst):
    unique = list(set(lst))
    if len(unique) < 2:
        return None
    for i in range(len(unique) - 1):
        for j in range(i + 1, len(unique)):
            if unique[i] > unique[j]:
                temp = unique[i]
                unique[i] = unique[j]
                unique[j] = temp
    return unique[-2]

print(second_largest(lst))
```

Problem 1: Weather Data Analyzer

You are given a list of daily average temperatures (in °C) for a month. Write a program that:

1. Reads n (number of days) and then n temperature values from the user.
2. Implements the following functions:
   · mean(temps) → returns the average temperature.
   · max_min(temps) → returns a tuple (max_temp, min_temp).
   · above_threshold(temps, threshold) → returns a list of temperatures strictly greater than threshold.
3. In the main part, call these functions to display:
   · Mean temperature (rounded to 2 decimals).
   · Highest and lowest temperature.
   · All temperatures above a user-specified threshold.
4. Use only basic data types, loops, lists, and functions (no external libraries like statistics or numpy).

```python
def mean(temps):
    total =0
    for i in temps:
        total+=i
    return total/len(temps)
    
def max_min(temps):
    max_temp = temps[0]
    min_temp = temps[0]
    for i in temps:
        if i>max_temp:
            max_temp=i
        if i<min_temp:
            min_temp=i
    return (max_temp,min_temp)
def above_threshold(temps,threshold):
    result = []
    for i in temps:
        if i>threshold:
            result.append(i)
    return result


n = int(input("Enter No. of days:"))
temps = []
for i in range(n):
    temp = float(input(f"Enter the temperature of Day:{i+1} :"))
    temps.append(temp)

threshold = float(input("Enter Threshold temperature:"))

avg = mean(temps)
max_temp = max_min(temps)
thres = above_threshold(temps,threshold)
print(f"Average Temperature is {avg}")
print(f"Max Temp:{max_temp[0]} and Min Temp:{max_temp[1]}")
print(f"Threshold temp:{thres}")
```

write a function that takes a list and return True if any value apperas at least twice , otherwise false use a set to track seen element

```python
lst = []
n = int(input("give how many element you want to put in list"))
for i in range(n):
    tmp = input(f"Enter {i+1} element:")
    lst.append(tmp)
    
def has_duplicates(lst):
    seen = set()
    
    for item in lst:
        print(f"Checking:{item}  Seen so far:{seen}")
        
        if item in seen:
             
            print(f"Duplicate found:{item}")
            return True
        
        seen.add(item)
    
    return False
    
data = has_duplicates(lst)
print(data)

```
