# Bubble Sort

```python
a = [1,2,3,8,5,6,2,8,9]
length = len(a)
print(a)
for i in range(length-1):
    swap = False
    for j in range(length - 1-i):
        if a[j]>a[j+1]:
            print(f"{a[j]} is swapped with {a[j+1]}")
            a[j],a[j+1]=a[j+1],a[j]
            print(a)
            swap=True
        else:
            print(f"{a[j]} is not swapped with {a[j+1]} because {a[j]} is smaller than {a[j+1]} ")
            
    print("")
    print(f"Pass {i} : {a}")
    print("")
    if swap==False:
        print("Completed")
        break
    print(f"Now Pass {i+1}")
    
        
```

# Insertion Sort


```python
lst =[9,4,5,3,8,1,6,7]
l = len(lst)
print(f"Unsorted list:{lst}")
for i in range(1,l):
    key = lst[i]
    j = i-1
    while j>=0 and lst[j] > key:
        print(f"Key:{key}")
        print(f"{lst[j+1]} swapped with {lst[j]}")
        lst[j+1] = lst[j]
        j=j-1
        print(lst)

    print("NO swapped")
    print(lst)
    lst[j+1] = key
    print(lst)
    

print(f"Sorted List{lst}")




```
