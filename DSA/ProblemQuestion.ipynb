Using  split() method.

```python

a = input("Give a comma-Separated Value like 1,2,3,4,5")

# print(type(a))
b =a.split(",")
# print(b)
x=[]
for i in b:
    x.append(int(i))
print(x)
sum =0
avg =0
for i in x:
    sum+=i
print(f"Sum:{sum}")
print(f"Average:{sum/len(x)}")
    
```

Q: vowel count lower case and upper case

```python
a= input("Enter your string to check vowel:")
count = 0
for i in a:
    if i == "a" or i == "A":
        count+=1
    elif i == "e" or i == "E":
        count+=1
    elif i == "i" or i == "I":
        count+=1
    elif i == "o" or i == "O":
        count+=1
    elif i == "u" or i == "U":
        count+=1    
print(f"Input:{a}")
print(f"No of vowel :{count}")
```

WAP to make calculator make function of operation

```python
# opr = input("Enter Operator (+,-,*,/)")
# opd1 = input("First number:")
# opd2 = input("Second number:")

def add(opd1,opd2):
    return opd1+opd2
def subtract(opd1,opd2):
    return opd1-opd2
def multiply(opd1,opd2):
    return opd1*opd2
def divide(opd1,opd2):
        return opd1/opd2


try:
    opr = input("Enter operator (+, -, *, /): ")
    opd1 = float(input("Enter first number: "))
    opd2 = float(input("Enter second number: "))
    

    if opr == "+":
        result = add(opd1, opd2)
    elif opr == "-":
        result = subtract(opd1, opd2)
    elif opr == "*":
        result = multiply(opd1, opd2)
    elif opr == "/":
        result = divide(opd1, opd2)
    else:
        result = "Invalid operator"
    print("Result:", result)

except ZeroDivisionError:
    print("Error: Can't divide by 0")

except ValueError:
    print("Error: Please enter valid numbers")
```

```python
t="anagram"
s="nagaram"
def isAnagram(s,t):
    if len(s) != len(t):
        return False
    d={}
    for i in s:
        if i in d:
            d[i]+=1
        else:
            d[i] =1
    for i in t:
        if i not in d:
            return False
        d[i]-=1
    for j in d.values():
        if j !=0:
            return False
    return True

print(isAnagram(t,s))
```

Prime no code


```python
x = [53,91,72,49,31,29]

def isPrime(x):
    for num in x:
        if num <= 1:
            print(f"{num} is not prime")
            continue
        
        for i in range(2, num):
            if num % i == 0:
                print(f"{num} is not prime")
                break
                
        else:
            print(f"{num} is a Prime No")

isPrime(x)
```
